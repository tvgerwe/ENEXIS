```python
import sqlite3
import pandas as pd

# Connect to the SQLite database
conn = sqlite3.connect('../Data/WARP.db')

# Read the raw_weather_preds table into a DataFrame
raw_weather_preds = pd.read_sql_query("SELECT * FROM raw_weather_preds", conn)

# Close the connection
conn.close()
```

```python
# Rename columns: replace 'previous_' with '-'
raw_weather_preds.columns = [col.replace('previous_day', '-') for col in raw_weather_preds.columns]

# Rename DataFrame variable
processed_weather_preds = raw_weather_preds

# Print overview of features
print("Missing values per column:")
print(processed_weather_preds.isnull().sum())
print("Columns after renaming:")
print(processed_weather_preds.columns.tolist())
print("\nDataFrame info:")
print(processed_weather_preds.info())

```

```python

# Store processed_weather_preds as a new table in WARP.db
conn = sqlite3.connect('../Data/WARP.db')
processed_weather_preds.to_sql('process_weather_preds', conn, if_exists='replace', index=False)
conn.close()

# Re-access the table and print features overview
conn = sqlite3.connect('../Data/WARP.db')
df_check = pd.read_sql_query("SELECT * FROM process_weather_preds", conn)
conn.close()

print("Date range:")
print(f"{df_check['date'].iloc[0]} to {df_check['date'].iloc[-1]}")
print("Columns in process_weather_preds:")
print(df_check.columns.tolist())
print("\nDataFrame info:")
print(df_check.info())
```
