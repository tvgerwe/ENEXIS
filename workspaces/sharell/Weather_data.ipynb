```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


# load the CSV file directly from GitHub using the raw URL.
github_url = "https://raw.githubusercontent.com/tvgerwe/ENEXIS/march-Twan/workspaces/twan/Weather_Data/Data/hist_weather_data_2022-2024_app_temp2025-03-19T09-46-18%2B00-00.csv"
df = pd.read_csv(github_url)
# Convert 'date' column to datetime
df['date'] = pd.to_datetime(df['date'])

print(df.head())
```

```python
# Select the variables of interest
variables = [
    "temperature_2m", "apparent_temperature", "cloud_cover", 
    "wind_speed_10m", "diffuse_radiation", "direct_normal_irradiance", 
    "shortwave_radiation", "wind_speed_100m"
]

# Create an empty dictionary to store correlation matrices
correlation_matrices = {}

# Iterate over each variable
for variable in variables:
    # Pivot the data to create a matrix with locations as columns and dates as rows
    pivot_df = df.pivot(index='date', columns='location', values=variable)
    
    # Calculate the correlation matrix for the current variable
    correlation_matrix = pivot_df.corr()
    
    # Store the correlation matrix in the dictionary
    correlation_matrices[variable] = correlation_matrix
    
    # Print the correlation matrix
    print(f"Correlation matrix for {variable}:")
    print(correlation_matrix)
    print("\n")
```
